```python
#imports
import numpy as np
import pandas as pd
import sys
import os
```

```python
#setting up the environment
DFdata = os.getenv("DFDATA")
print (DFdata)
```

```python
#url used: http://urbanprofiler.cloudapp.net/dataset/ia2d-e54m/
df_water = pd.read_csv(DFdata + '/ia2d-e54m/1414245804/ia2d-e54m')
```

```python
#previewing the dataset
df_water.head()
```

```python
#dropping a dataset column
df_water.drop(['Per Capita(Gallons per person per day)'], axis=1, inplace=True)
```

```python
#previewing the modified dataset
df_water.head()
```

```python
#plotting!
import pylab as pl
%pylab inline
pl.style.use('seaborn-darkgrid')
fig = pl.figure(figsize(16,8))

df_water.plot.scatter(x='Year' , y='NYC Consumption(Million gallons per day)')

pl.title("NYC Water Consumption Per Day - From 1979 to 2009", fontsize=20)
pl.xlabel("Year", fontsize=20)
pl.ylabel("NYC Consumption (Million gallons per day)", fontsize=20);
```
